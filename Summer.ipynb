### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
df[df['Medal'] == 'Gold']
```

```python
gmd=df[df['Medal'] == 'Gold']

sp=[]
for i in gmd['Sport'].unique():
  sp.append([i, len(gmd[gmd['Sport'] == i])])


pd.DataFrame(sp, columns = ['sport','medal']).sort_values(by = 'medal', ascending = False).head()
```

```python
pd.DataFrame(sp, columns = ['sport','Gmedal']).sort_values(by = 'Gmedal', ascending = False).head().plot(x = 'sport', y = 'Gmedal', kind = 'bar', figsize = (5,5))
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
gmd = df[df['Medal'] == 'Gold']
gi= gmd[gmd['Country'] == 'IND']

x=0
ig= []
for i in gi['Year'].unique():
     ig.append(i)
x=min(ig)

print("India won its first gold medal in",x)
```

### 7. Which event is most popular in terms on number of players? (Top 5)

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
gmd = df[df['Medal'] == 'Gold']
gw = gmd[gmd['Gender'] == 'Women']

women = []
for i in gw['Sport'].unique():
     women.append([i,len(gw[gw['Sport'] == i])])
 
pd.DataFrame(women, columns = ['Sport', 'gold medal']).sort_values(by = 'gold medal', ascending = False).head()
```

```python
pd.DataFrame(women, columns = ['Sport', 'gold medal']).sort_values(by = 'gold medal', ascending = False).head().plot(x = 'Sport', y = 'gold medal', kind = 'bar', figsize = (5,5))
```
